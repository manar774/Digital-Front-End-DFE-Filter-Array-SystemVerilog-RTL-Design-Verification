```python
import numpy as np

# ======================================================
# Load RTL and golden model outputs
# ======================================================
rtl_file    = "rtl_out_frac.csv"
golden_file = "frac_expected.txt"

rtl_out    = np.loadtxt(rtl_file, dtype=np.int32)
golden_out = np.loadtxt(golden_file, dtype=np.int32)

# ======================================================
# Align lengths (no skipping)
# ======================================================
N = min(len(golden_out), len(rtl_out))
gold_aligned = golden_out[:N]
rtl_aligned  = rtl_out[:N]

# ======================================================
# RMS calculations
# ======================================================
def rms(signal):
    return np.sqrt(np.mean(signal.astype(np.int64)**2))

rms_rtl    = rms(rtl_aligned)
rms_golden = rms(gold_aligned)

# RMS error (difference signal)
diff       = rtl_aligned.astype(np.int64) - gold_aligned.astype(np.int64)
rms_error  = rms(diff)

# Normalized error (relative to golden RMS)
norm_error = rms_error / rms_golden if rms_golden > 0 else float('nan')

# ======================================================
# Print results
# ======================================================
print("=== FRACTIONAL RMS Comparison (Full Output) ===")
print(f"Golden RMS   = {rms_golden:.2f}")
print(f"RTL RMS      = {rms_rtl:.2f}")
print(f"RMS Error    = {rms_error:.2f}")
print(f"Relative Err = {norm_error*100:.4f}%")
```
